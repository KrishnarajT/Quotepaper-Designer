```python
import cv2
import matplotlib.pyplot as plt

# main function
def detect_face(imagePath):
    # define the cascade clasifiers
    faceCascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml')
    eyeCascade = cv2.CascadeClassifier('data/haarcascade_eye.xml')  

    # Read the image
    image = cv2.imread(imagePath)
    # image = cv2.resize(image, (0, 0), fx=0.75, fy=0.75)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # plt.imshow(image)
    # plt.show()


    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE #flags = cv2.cv.CV_HAAR_SCALE_IMAGE
    )

    print("Found {0} faces!".format(len(faces)))


    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)



    # Detect eyes in the image
    eyes = eyeCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE #flags = cv2.cv.CV_HAAR_SCALE_IMAGE
    )

    print("Found {0} eyes!".format(len(eyes)))


    # Draw a rectangle around the faces
    for (x, y, w, h) in eyes:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)



    num=0
    if len(faces) > 0 and len(eyes) > 0:
        cv2.imwrite("image/modified%d.jpg" % i, image)
        # plt.imshow(image)
        # plt.show()
        return True
    return False
```

# Editing the Quote file to remove unwanted ones

```python
def modify_quotes(filename):
    with open(f'{filename}', 'r') as f:
        quote_list = f.read().split('%')
        for i in range(len(quote_list)):
            quote_lines = quote_list[i].split('\n')
            for j in range(len(quote_lines)):
                quote_line_words = quote_lines[j].split(' ')
                for l in range(len(quote_line_words)):
                    quote_line_words[l] = quote_line_words[l] + ' '
                for k in range(9, len(quote_line_words), 9):
                    quote_line_words[k] = quote_line_words[k] + '\n'

                tmp = ''
                tmp = tmp.join(quote_line_words)
                quote_lines[j] = tmp
            for m in range(len(quote_lines)):
                quote_lines[m] = quote_lines[m] + '\n'
            quote_list[i] = ''
            quote_list[i] = quote_list[i].join(quote_lines)
            quote_list[i] = quote_list[i].strip('\n')
        for i in quote_list[:]:
            if i.count('\n') > 10:
                quote_list.remove(i)
    with open(f'Finished Quotes/modified{filename.name}', 'w') as fout:
        bulkwrite = ''
        for i in range(len(quote_list)):
            quote_list[i] = quote_list[i] + '%'
        bulkwrite = bulkwrite.join(quote_list)   
        fout.write(bulkwrite)
        
```

```python
def find_timestamp(all_subs, i):
    tmp_list = all_subs[i-4:i]
    for _, j in enumerate(tmp_list):
        if '-->' in j:
            break
    print(tmp_list[_])
    timestamp = tmp_list[_].split(',')[0].split(':')
    timestamp.append(int(tmp_list[_].split(',')[1][:3]))
    # timestamp = tmp_list[_].split('>')[1].strip(' ').split(',')[0].split(':')
    for _, i in enumerate(timestamp):
        timestamp[_] = int(timestamp[_])
    timestamp[2] = timestamp[2] + 3
    return timestamp
```

```python
def search(path_to_srt, text):
    with open(path_to_srt, 'r', errors = 'ignore') as f:
        all_subs = f.readlines()

    for _, i in enumerate(all_subs):
        if text.lower() in i.lower():
            print(_, i)
            found = _
            break
        else: found = 0
    return found
        
```
